```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro

def sqrt_transformation(Df, column_name):
    Df[f'{column_name}_sqrt'] = np.sqrt(Df[column_name])
    
    stat, p_value = shapiro(Df[f'{column_name}_sqrt'])
    
    sns.kdeplot(Df[f'{column_name}_sqrt'], fill=True)
    plt.title(f'Distribution of {column_name}_sqrt')
    plt.xlabel(f'{column_name}_sqrt')
    plt.ylabel('Density')
    plt.show()
    
    print('Shapiro-Wilk Test p-value:', p_value)

# Load your Excel file
df = pd.read_excel(r"C:\Users\ACER\Desktop\Udaanous\Python\Python PDF\boxplot_data.xlsx")

# Call the function with the actual column name
sqrt_transformation(df, 'Sales Revenue Generated')
```

```python
def log_transformation(data, column_name):
    data[f"{column_name}_log"] = np.log(data[column_name])
    stat, p_value = shapiro(data[f"{column_name}_log"])
    distribution = sns.kdeplot(data[f"{column_name}_log"])
    return distribution, p_value
# Load your Excel file
df = pd.read_excel(r"C:\Users\ACER\Desktop\Udaanous\Python\Python PDF\boxplot_data.xlsx")

# Call the function with the actual column name
sqrt_transformation(df, 'Days Present')
```
