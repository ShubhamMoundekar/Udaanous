```python
#Create a correlation matrix using the variables:

numeric_cols = ['Task Completion Rates', 'Decision-Making Skills Rating', 'Sales Revenue Generated']
correlation_matrix = Data1[numeric_cols].corr().round(3)
plt.figure(figsize=(6,8))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Heatmap with Rounded Values")
plt.show()
```

# Task 2 Using Boxplot Data

```python
Box_plot = pd.read_excel(r"C:\Users\ACER\Desktop\Udaanous\Python\Python PDF\boxplot_data.xlsx")
Df = pd.DataFrame(Box_plot)
Df.head(5)
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(x="churned",y="Task Completion Rates",data=Df)
plt.title("Boxplot of Task Completion Rates by Churned Status")
plt.show()
```

```python
median_churned = Df[Df["churned"] == "Yes"]["Task Completion Rates"].median()
print("Median Task Completion Rate for churned employees:", median_churned)
```
