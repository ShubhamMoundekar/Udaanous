# Pythom Sale Mini Project 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\ACER\Downloads\Sale.csv")
df
```

```python
# Convert OrderDate to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
```

```python
# Extract month from date
df['Month'] = df['OrderDate'].dt.month
```

```python
# Top products
top_products = df.groupby('ProductName')['Sales'].sum().sort_values(ascending=False)
```

```python
# Monthly sales
monthly_sales = df.groupby('Month')['Sales'].sum()
```

```python
# Regional sales
regional_sales = df.groupby('Region')['Sales'].sum()
```

```python
# Print outputs
print("\nüì¶ Top Products by Sales:")
print(top_products)

print("\nüìÖ Monthly Sales:")
print(monthly_sales)

print("\nüåç Regional Sales:")
print(regional_sales)
```

# üìä Bar Plot

```python
plt.figure(figsize=(8, 6))
top_products.head(10).plot(kind='bar', color='skyblue', title='Top 10 Products by Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# üìà Line Plot

```python
monthly_sales.plot(kind='line', marker='o', color='orange', title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
```

# üìä Bar Plot

```python
regional_sales.plot(kind='bar', color='green', title='Sales by Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# üìå 4. Summary Insights

```python
print("\nüìà Summary Insights:")
print(f"üî∏ Highest-selling product: {top_products.idxmax()}")
print(f"üî∏ Month with highest sales: {monthly_sales.idxmax()}")
print(f"üî∏ Top-performing region: {regional_sales.idxmax()}")
```
