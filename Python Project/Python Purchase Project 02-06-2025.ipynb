# Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\ACER\Desktop\Udaanous\Python\Python Project\customer_analysis_project\data\customers.csv")

# Display first few rows
df
```

# Data Exploration

```python
# Check for missing values
print(df.isnull().sum())

# Statistical summary
print(df.describe())

# Data types and structure
df.info()
```

# Visualize Gender Distribution

```python
# Count of customers by gender
sns.countplot(data=df, x='Gender', palette='Set2')

# Add labels and title
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

# Analyze Spending Score by Gender

```python
# Boxplot to compare Spending Score across Genders
sns.boxplot(data=df, x='Gender', y='SpendingScore', palette='pastel')

# Add labels and title
plt.title("Spending Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

# Relationship Between Age and Spending Score

```python
# Scatter plot to show relationship between Age and Spending Score
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='SpendingScore', hue='Gender', palette='coolwarm')

# Add labels and title
plt.title("Age vs SpendingScore")
plt.xlabel("Age")
plt.ylabel("SpendingScore")
plt.legend(title='Gender')
plt.show()
```

# Category-Wise Spending Analysis

```python
# Check if the expected columns exist
if 'Category' in df.columns and 'Amount Spent' in df.columns:
    # Group by Category and calculate average spending
    category_avg = df.groupby('Category')['Amount Spent'].mean().sort_values(ascending=False)

    # Bar plot of average spending by category
    category_avg.plot(kind='bar', color='skyblue')
    plt.title("Average Spending by Category")
    plt.xlabel("Purchase Category")
    plt.ylabel("Average Amount Spent")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'Category' and/or 'Amount Spent' not found in the dataset.")
```

# Optional Enhancements

```python
# Income vs Spending Score
if 'Annual Income (k$)' in df.columns:
    sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender')
    plt.title("Income vs Spending Score")
    plt.show()

# Pie chart for Gender
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```
